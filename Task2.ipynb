```python
#importing libraries with alias
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
#Loading Dataset
df=pd.read_csv('Titanic.csv')
df
```

```python
#displaying first five elements
df.head()
```

```python
#Displaying last five elements
df.tail()
```

```python
#Getting information about the loaded dataset
df.info()
```

```python
#Getting Statistical Summary of the Numerical Columns
df.describe()
```

```python
#Getting the count of null values in each column
df.isnull().sum()
```

```python
#Deleting the most inconsistent column i.e, having more NULL values and is not useful
df=df.drop('Cabin',axis=1)
df
```

```python
df.columns
```

```python
#calculating some statistical terms
mean_age=df['Age'].mean()
median_age=df['Age'].median()
mode_age=df['Age'].mode()
```

```python
print(mean_age)
print(median_age)
print(mode_age)
```

```python
#Handling Missing values
df['Age']=df['Age'].fillna(df['Age'].median())
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode())
```

```python
#knowing the unique elements in a column
df['Sex'].value_counts()
```

```python
df['Embarked'].value_counts()
```

```python
#Now encoding of Non-Numerical Columns
df['Sex']=df['Sex'].replace('male',0)
df['Sex']=df['Sex'].replace('female',1)
```

```python
df['Embarked']=df['Embarked'].map({'C':0,'Q':1,'S':2})
```

```python
df
```

```python
#Visualizing Survival count using Barplot
sn.countplot(x='Survived', data=df)
plt.title('Survival Count (0 = No, 1 = Yes)')
plt.show()
```

```python
#visualizing ages using histogram
plt.hist(df['Age'], bins=100, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#Visualizing survival rate based on gender
sn.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
#Checking whether Higher fares are correlated using Boxplot
sn.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare Distribution by Survival (0-No,1-Yes)')
plt.show()
```

```python
# Select numerical columns for correlation
corr_df = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex','Embarked']]

# Plot correlation matrix
plt.figure(figsize=(8, 6))
sn.heatmap(corr_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
